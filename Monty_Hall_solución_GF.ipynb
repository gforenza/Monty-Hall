**Solución problema Monty Hall (Gabriel Forenza)**

```
import random
# Se corren 1000 simulaciones
N= 1000
# Se crean dos listas. La lista cambios guarda los resultados de cambiar de puerta. La lista no_cambios guarda los resultados de no cambiar de puerta
no_cambios = []
cambios = []
for i in range(N):
    # Se crea la lista puertas con un auto y dos cabras.
    puertas = ['auto', 'cabra', 'cabra']
    # Se aleatorizan las puertas
    random.shuffle(puertas)
    # Se elige una puerta al azar
    indice_puerta_elegida = random.choice([0, 1, 2])
    # Se guarda el valor de la puerta elegida que sería el resultado si no cambio de puerta
    no_cambio = puertas.pop(indice_puerta_elegida)
    # Se quita una cabra y lo que queda en puertas sería el resultado de cambiar de puerta
    puertas.remove('cabra')
    cambio = puertas[0]
    # Se acumulan en listas los resultados de cambiar de puerta y de no cambiar de puerta
    no_cambios.append(no_cambio)
    cambios.append(cambio)
```

**Probabilidades de acierto:**

Sin cambiar la puerta elegida:

```
no_cambios.count('auto') / N
```

Cambiando la puerta elegida:

```
cambios.count('auto') / N
```

**Gráfico de hits en función del número de iteración:**

```
sumador1=0
sumador2=0
cambios_num= []
no_cambios_num= []
for i in range (N):
  if cambios[i]=='auto':
    sumador1=sumador1+1
  cambios_num.append(sumador1)
  if no_cambios[i]=='auto':
    sumador2=sumador2+1
  no_cambios_num.append(sumador2)

```

```
from matplotlib import pyplot as plt
```

```
plt.figure(figsize=(16,8))
plt.plot(no_cambios_num, label='No cambio')
plt.plot(cambios_num, label='Cambio')
plt.xlabel('Iteración')
plt.ylabel('hits')
plt.legend()
plt.show()
```
